## CORRELATION TESTING USING HEATMAP

```python
disease_df_1= disease_df[['Countries', 'Year', 'Meningitis', 'Neonatal disorders', 'Malaria', 'Maternal disorders', 'Cardiovascular diseases', 'Diarrheal diseases', 'Digestive diseases', 'Poisonings', 'Acute hepatitis']]
```

```python
disease_df_1
```

```python
disease_df_1.corr(method= 'pearson')
```

```python
correlation_matrix= disease_df_1.corr(method='spearman')
```

```python
sns.heatmap(correlation_matrix, annot= True )
plt.title("CORRELATION MATRIC FOR NUMERIC FEATURE")
plt.xlabel('HEALTH FEATURE')
plt.ylabel('HEALTH FEATURE')
plt.show()
```
